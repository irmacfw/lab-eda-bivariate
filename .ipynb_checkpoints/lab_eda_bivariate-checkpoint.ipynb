```python
# Part 1: Best-Seller Trends Across Categories
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
import numpy as np

%matplotlib inline

df = pd.read_csv(r"C:\Users\Gustavo\Documents\Irma\Ironhack\SEMANA 5\DAY 1\lab-eda-univariate\amz_uk_price_prediction_dataset.csv")
df.head()

```

```python
# 1 Crosstab
crosstab = pd.crosstab(df['category'], df['isBestSeller'])
crosstab
```

```python
#1.1 Proportion of Best-Sellers
prop_best = (crosstab.T / crosstab.sum(axis=1)).T
prop_best = prop_best.sort_values(by=True, ascending=False)  # True = isBestSeller
prop_best.head()
```

```python
# 2. Chi-square test
chi2, p, dof, expected = chi2_contingency(crosstab)
print("Chi-square:", chi2, "p-value:", p)
```

```python
# 3. Cramér’s V
n = crosstab.sum().sum()
cramers_v = np.sqrt(chi2 / (n * (min(crosstab.shape)-1)))
print("Cramer's V:", cramers_v)
```

```python
# 4. Visualizations
crosstab.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Best-seller distribution across categories")
plt.show()
```

```python
# Part 2: Prices and Ratings
```

```python
# Remove Outliers with IQR
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lb = Q1 - 1.5*IQR
ub = Q3 + 1.5*IQR

df_no_out = df[(df['price'] >= lb) & (df['price'] <= ub)]
```

```python
# Violin plot (top 20 categories by count)
top20 = df_no_out['category'].value_counts().head(20).index
sns.violinplot(data=df_no_out[df_no_out['category'].isin(top20)],
               x='category', y='price')
plt.xticks(rotation=90)
plt.title("Price distribution across top 20 categories")
plt.show()
```

```python
# Bar chart (avg price for top 10 categories)
top10 = df_no_out['category'].value_counts().head(10).index
avg_price = (df_no_out[df_no_out['category'].isin(top10)]
             .groupby('category')['price']
             .mean()
             .sort_values(ascending=False))

avg_price.plot(kind='bar')
plt.title("Average price by top 10 categories")
plt.ylabel("Average price")
plt.show()

```

```python
#Boxplot of Ratings
top10 = df_no_out['category'].value_counts().head(10).index
sns.boxplot(data=df_no_out[df_no_out['category'].isin(top10)],
            x='category', y='stars')
plt.xticks(rotation=90)
plt.title("Ratings distribution across top 10 categories")
plt.show()
```

```python
#Part 3: Prices and Ratings
```

```python
#Correlation
corr = df_no_out[['price','stars']].corr()
print(corr)
```

```python
#Scatter Plot
sns.scatterplot(data=df_no_out, x='stars', y='price', alpha=0.5)
plt.title("Price vs Stars")
plt.show()
```

```python
#Heatmap
num_df = df_no_out.select_dtypes(include='number')
sns.heatmap(num_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation heatmap (numeric columns only)")
plt.show()
```

```python
# QQ Plot
import scipy.stats as stats
import matplotlib.pyplot as plt

stats.probplot(df_no_out['price'], dist="norm", plot=plt)
plt.title("QQ plot of price")
plt.show()
```

Business Report – Amazon UK Product Insights
Part 1: Best-Seller Trends

Categories like Grocery and Smart Home Security & Lighting show higher proportions of best-sellers.

Chi-square test (p=0.0): best-seller status depends on category.

Cramér’s V (0.12): weak association → category matters but is not the only driver.

Implication: Promotions and investments should be adapted by category, but other factors (price, reviews, brand) are also key.

Part 2: Prices and Ratings

Prices: Premium categories include Handmade Home & Kitchen Products and Sports & Outdoors.

Ratings: Most categories maintain high ratings (close to 5), with slight variation in Handmade Gifts and Birthday Gifts.

Implication: Premium pricing is viable in strong-rating categories. Lower-rated ones may need quality improvements.

Part 3: Prices vs Ratings

Correlation (≈0.07): very weak → price does not affect ratings.

Scatter plot & heatmap: confirm independence of price and ratings.

QQ plot: price distribution is skewed, not normal.

Implication: Prices can be adjusted without major impact on ratings. Pricing models should account for outliers.

✅ Overall Insight:
Amazon UK can leverage premium pricing in high-rating categories and focus on categories with higher best-seller ratios, while monitoring quality in lower-rated segments.

Bonus: With Outliers

When keeping outliers, price distributions become highly skewed, with extreme values pulling up the averages.

Violin/box plots: harder to see the central tendency because a few very expensive products distort the scale.

Correlation results: still weak between price and ratings, but extreme prices exaggerate the spread.

QQ plot: confirms stronger deviation from normality.

Insight: Outliers mask the true behavior of most products. Removing them gives clearer, business-relevant insights, while keeping them highlights the existence of niche ultra-premium items.
